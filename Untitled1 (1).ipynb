Importing the required libraries



```python
# Importing all libraries required
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

## Collecting and Reading The Data

```python
# Reading data
url ='https://raw.githubusercontent.com/AdiPersonalWorks/Random/master/student_scores%20-%20student_scores.csv'
df= pd.read_csv(url)
print("Data imported successfully")
df.head(10)
```

## Plotting and visualizing the Data

```python
# PLotting the graph of scores
df.plot(x='Hours',y='Scores',style='.',color='green')
plt.title('Hours vs Scores')
plt.xlabel('Hours Studied')
plt.ylabel('Percentage score')
plt.show()
```


 **the above graph depicts that there is a positive correlation between hours and percentage**

## Observing The Distribution Of the Continuous Variables

```python
df.hist(['Hours','Scores'],figsize=(18,5))
```

```python
round(df.describe(),2)
```

```python
df.nunique()
```

## Checking Missing Values

```python
df.isnull().sum()
```

**Preparing The Data**
## The next step is to divide the data into "attributes" (inputs) and "labels" (outputs).

```python
a=df.iloc[:,:-1].values #all rows and columns except last column
b=df.iloc[:,1].values  #all rows and only 1st column i.e scores
```

```python
a
```

```python
b
```

## Splitting The Data Into Training and Testing Data

```python
from sklearn.model_selection import train_test_split
a_train,a_test,b_train,b_test = train_test_split(a,b,test_size=0.2,random_state=42000)
```

## Training the Algorithm 

```python
from sklearn.linear_model import LinearRegression  
regressor = LinearRegression()  
regressor.fit(a_train, b_train) 

print("Training complete.")
```

## Plotting The Regression Line

```python
# Plotting the regression line
line = regressor.coef_*a+regressor.intercept_

# Plotting for the test data
plt.scatter(a, b)
plt.plot(a, line);
plt.show()
```

## prediction on test data

```python
print(a_test) # Testing data - In Hours
b_pred = regressor.predict(a_test) # Predicting the scores
```

```python
# Comparing Actual vs Predicted
abss = pd.DataFrame({'Actual': b_test, 'Predicted': b_pred})  
abss
```

## graph between actual vs predicted value

```python

abss.plot(kind='bar',figsize=(5,5))
plt.grid(which='major',linewidth='0.5',color='blue')
plt.grid(which='minor',linewidth='0.5',color='red')
plt.show()
```

## Prediction of score when a student studies for 9.25 hours

```python
#test with the required data
Hours = [[9.25]] #predict takes 2D array 
own_pred = regressor.predict(Hours)
print("No of Hours studied = {}".format(Hours))
print("Predicted Score = {}".format(own_pred[0]))
```

```python
from sklearn import metrics  
print('Mean Absolute Error:', 
      metrics.mean_absolute_error(b_test, b_pred))
```

## Conclusion: When a student studies for 9.25 hours per day , the prediction score will be 94%
